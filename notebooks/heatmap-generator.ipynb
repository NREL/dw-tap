### This notebook saves geojson files with points for heatmap analysis into `../heatmaps`

```python
# Hide warnings
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import geopandas as gpd
from dw_tap.heatmap import gen_heatmap_points
%config InlineBackend.figure_format = 'svg'
```

```python
obstacle_file ="../sites/180-5BuildingsManual.geojson"
obstacles = gpd.read_file(obstacle_file)

n_points = 40
gridpoints = gen_heatmap_points(obstacles, 
                                n_points=n_points,
                                visualize=True, 
                                file_dest="../heatmaps/180-%dpoints.json" % (n_points*n_points))
```
