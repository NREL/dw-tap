```python
# This is a minimal working example of running QUIC on Eagle;
# QUIC files need to be accessible at the path referenced below (`quic_code_dir`)
# and the `quic_exec` file should have code for running QUIC with multiprocessing
# (where input gets split into `n_chunks` that are processed in parallel).

import os
import subprocess

quic_code_dir ="/campaign/tap/QUIC_TAP_API"
quic_exec = "run_quic_mp_v2.py"
script_path = os.path.join(quic_code_dir, quic_exec)

cwd = os.getcwd()

# Must use subprocess as a way of wrapping/calling the python script becuase that script uses multiprocessing
# and other methods seem to break
subprocess.run(["python", script_path,
                "--atmospheric_file", os.path.join(quic_code_dir, "180_2007_binned.csv"),
                "--obstacles_file", os.path.join(quic_code_dir, "180-5BuildingsManual.geojson"),
                "--lat", "39.32856",
                "--lon", "-89.40238",
                "--z_turbine", "15.0",
                "--n_chunks", "36",
                "--output", os.path.join(cwd, "example_39.32856_-89.40238.csv")],
               stdout=subprocess.DEVNULL, # This addition suppresses entire (lengthy) output
               cwd=quic_code_dir) # This addition is needed to access QUIC executables inside quic_code_dir directory
```
